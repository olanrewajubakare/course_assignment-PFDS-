```python
# Loading the training datasets
train_dataset = '../input/random-linear-regression/train.csv'
df_train_dataset = pd.read_csv(train_dataset)
df_train_dataset = df_train_dataset.dropna()
X_train = np.array(df_train_dataset.iloc[:, :-1].values)
y_train = np.array(df_train_dataset.iloc[:, 1].values)
print(y_train[0:5],'\n', X_train[0:5])
```

```python
# Loading the test datasets
testing_dataset = '../input/random-linear-regression/test.csv'
df_testing_dataset = pd.read_csv(testing_dataset)
df_testing_dataset = df_testing_dataset.dropna()
X_test = np.array(df_testing_dataset.iloc[:, :-1].values)
y_test = np.array(df_testing_dataset.iloc[:, 1].values)
print(X_test[0:5], '\n', y_test[0:5])
```

```python
import matplotlib.pyplot as plt 
fig = plt.figure(figsize = (10, 7))
plt.scatter(X_train, y_train, color = 'b')
plt.title('Relationship between the input and the output of the training dataset')
plt.xlabel('X_train')
plt.ylabel('y_train');
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print(lr.predict(X_test)[0:5])
print(y_test[0:5])
```

```python
accuracy = lr.score(X_test, y_test)
print('Accuracy = ' + str(accuracy))
```

```python
fig = plt.figure(figsize = (10, 7))
plt.style.use('seaborn')
plt.scatter(X_test, y_test, color = 'red', marker = 'o', s = 35, alpha = 0.5, 
            label = 'Test data')
plt.scatter(X_test, lr.predict(X_test), color = 'b', 
            label = 'Model plot') 
plt.title('Predicted Values vs Actual Values')
plt.xlabel('Inputs')
plt.ylabel('Actual Values')
plt.legend(loc = 2);
```
